```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path='/content/archive.zip'
df=pd.read_csv(file_path)
print(df.head())
```

```python
head=np.array(df['Head Size(cm^3)'],dtype=float)
brain=np.array(df['Brain Weight(grams)'],dtype=float)
n=len(head)
n
```

```python
sum_head = np.sum(head)
sum_brain = np.sum(brain)
sum_head_brain = np.sum(head * brain)
sum_head_square = np.sum(head * head)
```

```python
n = len(head)
mean_x = sum_head / n
mean_y = sum_brain / n
mean_xy = sum_head_brain / n
mean_x2 = sum_head_square / n
```

```python

w1 = (mean_xy - mean_x * mean_y) / (mean_x2 - (mean_x ** 2))
w0 = mean_y - w1 * mean_x
print(w1)
print(w0)
```

```python
y=w0+w1*head
print(y)
```

```python
plt.scatter(head,brain,color='g',label='Actual Data')
plt.plot(head,y,color='red',label='Regression line')
plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
plt.legend()
plt.show()
```

```python
meansquare=np.mean((brain-y)**2)
print(meansquare)
```
